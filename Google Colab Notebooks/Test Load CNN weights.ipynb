```python
from EEGArtifactRemoval.models import UnidimensionalResCNN

model = UnidimensionalResCNN(400)
model.load('/content/drive/MyDrive/1D-ResCNN.h5')
```

```python
import numpy as np
import matplotlib.pyplot as plt
test_data = np.load(open('/content/drive/MyDrive/test_data.pickle', 'rb'),allow_pickle=True)
test_labels = np.load(open('/content/drive/MyDrive/test_labels.pickle', 'rb'),allow_pickle=True)
```

```python
result = model.evaluate(test_data)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharey=True)
plt.ylim(-2.1,5.7)

index=265

fig.set_size_inches(25, 9)

ax1.plot(test_labels[index], label="Contaminated")
ax1.grid()

ax2.plot(test_data[index], label='Prediction')
ax2.grid()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharey=True)
plt.ylim(-2,5.7)


index=265

fig.set_size_inches(25, 9)

ax1.plot(test_data[index], label="Contaminated")
ax1.grid()

ax2.plot(result[index].flatten(), label='Prediction')
ax2.grid()

plt.show()
```
